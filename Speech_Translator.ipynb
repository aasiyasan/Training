# Importing Packages

```python
import speech_recognition as sr
from gtts import gTTS
import playsound
import translators as tr
import os 
#pyaudio needs to be installed
```

# Converting Input Speech To Text

```python
# It requires an internet connection, since the recognition is cloud-based.

# English
def hugo_listen_en():
    # Creating a recognizer
    r = sr.Recognizer()

    with sr.Microphone() as source :
        # this listens to what is said in the microphone 
        audio = r.listen(source)
        text = r.recognize_google(audio,language = 'en-IN')
    
    text = text.lower()
    return text

#Telugu
def hugo_listen_te():
    # Creating a recognizer
    r = sr.Recognizer()

    with sr.Microphone() as source :
        # this listens to what is said in the microphone 
        audio = r.listen(source)
        text = r.recognize_google(audio,language = 'te-IN')
    
    text = text.lower()
    return text

#Tamil
def hugo_listen_ta():
    # Creating a recognizer
    r = sr.Recognizer()

    with sr.Microphone() as source :
        # this listens to what is said in the microphone 
        audio = r.listen(source)
        text = r.recognize_google(audio,language = 'ta-IN')
    
    text = text.lower()
    return text

#Hindi
def hugo_listen_hi():
    # Creating a recognizer
    r = sr.Recognizer()

    with sr.Microphone() as source :
        # this listens to what is said in the microphone 
        audio = r.listen(source)
        text = r.recognize_google(audio,language = 'hi-IN')
    
    text = text.lower()
    return text

#Kanada
def hugo_listen_kn():
    # Creating a recognizer
    r = sr.Recognizer()

    with sr.Microphone() as source :
        # this listens to what is said in the microphone 
        audio = r.listen(source)
        text = r.recognize_google(audio,language = 'kn-IN')
    
    text = text.lower()
    return text
```

# Coverting the Text to Speech

```python
#English
def hugo_speak_en(text):
    # Creating an audio file
    file_name = "audio_data.mp3"

    # Converting Text to Speech
    tts = gTTS(text = text, lang= "en")

    # Save File
    tts.save(file_name)

    # Playing file
    playsound.playsound(file_name)

    # Remove File
    os.remove(file_name)

#Hindi
def hugo_speak_hi(text):
    # Creating an audio file
    file_name = "audio_data.mp3"

    # Converting Text to Speech
    tts = gTTS(text = text, lang= "hi")

    # Save File
    tts.save(file_name)

    # Playing file
    playsound.playsound(file_name)

    # Remove File
    os.remove(file_name)

#Telugu
def hugo_speak_te(text):
    # Creating an audio file
    file_name = "audio_data.mp3"

    # Converting Text to Speech
    tts = gTTS(text = text, lang= "te")

    # Save File
    tts.save(file_name)

    # Playing file
    playsound.playsound(file_name)

    # Remove File
    os.remove(file_name)

#Tamil
def hugo_speak_ta(text):
    # Creating an audio file
    file_name = "audio_data.mp3"

    # Converting Text to Speech
    tts = gTTS(text = text, lang= "ta")

    # Save File
    tts.save(file_name)

    # Playing file
    playsound.playsound(file_name)

    # Remove File
    os.remove(file_name)

#Kanada
def hugo_speak_kn(text):
    # Creating an audio file
    file_name = "audio_data.mp3"

    # Converting Text to Speech
    tts = gTTS(text = text, lang= "kn")

    # Save File
    tts.save(file_name)

    # Playing file
    playsound.playsound(file_name)

    # Remove File
    os.remove(file_name)
```

# Translating From One Language To Another

```python
# English - Hindi 
def translate_en_hi(text):
    hugo_speak_hi(tr.translate_text(text ,from_language = 'en' ,to_language = 'hi'))

# English - Telugu 
def translate_en_te(text):
    hugo_speak_te(tr.translate_text(text ,from_language = 'en' ,to_language = 'te'))

# English - Tamil
def translate_en_ta(text):
    hugo_speak_ta(tr.translate_text(text ,from_language = 'en' ,to_language = 'ta'))

# English - Kanada 
def translate_en_kn(text):
    hugo_speak_kn(tr.translate_text(text ,from_language = 'en' ,to_language = 'kn'))


# Hindi - English
def translate_hi_en(text):
    hugo_speak_en(tr.translate_text(text ,from_language = 'hi' ,to_language = 'en'))

# Hindi - Telugu
def translate_hi_te(text):
    hugo_speak_te(tr.translate_text(text ,from_language = 'hi' ,to_language = 'te'))

# Hindi - Tamil
def translate_hi_ta(text):
    hugo_speak_ta(tr.translate_text(text ,from_language = 'hi' ,to_language = 'ta'))

# Hindi - Kanada
def translate_hi_kn(text):
    hugo_speak_kn(tr.translate_text(text ,from_language = 'hi' ,to_language = 'kn'))


# Telugu - English
def translate_te_en(text):
    hugo_speak_en(tr.translate_text(text ,from_language = 'te' ,to_language = 'en'))

# Telugu - Hindi
def translate_te_hi(text):
    hugo_speak_hi(tr.translate_text(text ,from_language = 'te' ,to_language = 'hi'))

# Telugu - Tamil
def translate_te_ta(text):
    hugo_speak_ta(tr.translate_text(text ,from_language = 'te' ,to_language = 'ta'))

# Telugu - Kanada
def translate_te_kn(text):
    hugo_speak_kn(tr.translate_text(text ,from_language = 'te' ,to_language = 'kn'))


# Tamil - English
def translate_ta_en(text):
    hugo_speak_en(tr.translate_text(text ,from_language = 'ta' ,to_language = 'en'))

# Tamil - Hindi
def translate_ta_hi(text):
    hugo_speak_hi(tr.translate_text(text ,from_language = 'ta' ,to_language = 'hi'))

# Tamil - Telugu
def translate_ta_te(text):
    hugo_speak_te(tr.translate_text(text ,from_language = 'ta' ,to_language = 'te'))

# Tamil - Kanada
def translate_ta_kn(text):
    hugo_speak_kn(tr.translate_text(text ,from_language = 'ta' ,to_language = 'kn'))


# Kanada - English
def translate_kn_en(text):
    hugo_speak_en(tr.translate_text(text ,from_language = 'ka' ,to_language = 'en'))

# Kanada - Hindi
def translate_kn_hi(text):
    hugo_speak_hi(tr.translate_text(text ,from_language = 'kn' ,to_language = 'hi'))

# Kanada - Telugu
def translate_kn_te(text):
    hugo_speak_te(tr.translate_text(text ,from_language = 'kn' ,to_language = 'te'))

# Kanada - Tamil
def translate_kn_ta(text):
    hugo_speak_kn(tr.translate_text(text ,from_language = 'kn' ,to_language = 'ta'))
```

# Running The Assistant

```python
def hugo_run():

    hugo_speak_en("Hey I am Hugo , Your translator assistant ! What's your name ?")
    name = hugo_listen_en()
    hugo_speak_en(f"Hey {name} ! Tell me what do you want to translate ?")
    listen_assistant = hugo_listen_en()
    print(listen_assistant)
    hugo_reply(listen_assistant)

```
